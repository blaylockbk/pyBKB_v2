```python
# Rhea Fire
lat = 35.849
lon = -99.213
```

```python
# Create Map Object
mapCenterLat = lat
mapCenterLon = lon
m = Basemap(projection='cyl', resolution='h', area_thresh=1000,
            llcrnrlon=mapCenterLon-5.4,  urcrnrlon=mapCenterLon+5.6,
            llcrnrlat=mapCenterLat-3.4, urcrnrlat=mapCenterLat+3.4)
```

```python
#m.drawcounties()
m.drawstates()
m.arcgisimage(service='World_Shaded_Relief', xpixels=1000, dpi=1000)
```

```python
DATE = datetime(2018, 4, 12, 23, 42)
G = get_GOES16_truecolor(file_nearest(DATE))
F = get_GOES16_firetemperature(file_nearest(DATE))
```

```python
newmap = m.pcolormesh(G['lon'], G['lat'], G['TrueColor'][:,:,1],
                      color=G['rgb_tuple'],
                      linewidth=0)
newmap.set_array(None)
m.drawstates(color='w', linewidth=.25)
```

```python
newmap2 = m.pcolormesh(F['lon'], F['lat'], F['TrueColor'][:,:,1],
                      color=F['rgb_tuple'],
                      linewidth=0)
newmap2.set_array(None)
m.drawstates(color='w', linewidth=.25)
```

## Blend the Fire Temperature and the True Color images
Blend by taking the max RGB value for each pixel

```python
max_RGB = np.nanmax([F['rgb_tuple'], G['rgb_tuple']], axis=0)
```

```python
newmap5 = m.pcolormesh(G['lon'], G['lat'], G['TrueColor'][:,:,1],
                      color=max_RGB,
                      linewidth=0)
newmap5.set_array(None)

m.drawstates(color='w', linewidth=.5)
plt.savefig('./figs/Fig_08_GOES_TrueColor_FireTemp')
```
